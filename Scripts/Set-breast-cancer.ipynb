```python
import pandas as pd
import os
random_state = 7
```

```python
path = os.getcwd()
parent = os.path.dirname(path)

df = pd.read_csv(parent + '\Data\\' + "datasets123_allinfo" + ".csv")


```

```python
df.head()
```

```python
print(df["Disease"].value_counts())
print()
```

```python
df = df.loc[df['Disease'].isin(['healthy','Breast cancer'])]
```

```python
print(df["Disease"].value_counts())
print()
```

```python
df['target'] = [1 if (x == 'Breast cancer') else 0 for x in df['Disease']]
```

```python
df.head(30)
```

```python
print(df["target"].value_counts())
print()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.2, random_state=random_state, stratify=df[['target']])
```

```python
train.head()
```

```python
test.head()
```

```python
print(train["target"].value_counts())
print()
print(test["target"].value_counts())
print()
```

```python
train = train.drop(["Dataset", "GSE", "Condition", "Disease", "Tissue", "FAB", "Filename", "id"], axis = 1)
```

```python
test = test.drop(["Dataset", "GSE", "Condition", "Disease", "Tissue", "FAB", "Filename", "id"], axis = 1)
```

```python
train.to_csv("trainset_breast_healthy_80_target_in.csv" ,index=False)
```

```python
test.to_csv("testset_breast_healthy_20_target_in.csv" ,index=False)
```
