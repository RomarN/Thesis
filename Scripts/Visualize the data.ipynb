```python
import pandas as pd
import seaborn as sns
import os
import matplotlib.pyplot as plt
```

```python
path = os.getcwd()
parent = os.path.dirname(path)

df = pd.read_csv(parent + "/Data/datasets123_allinfo.csv")
```

```python
df.describe()
```

```python
df
```

```python
disease_freq = df.Disease.value_counts()
```

```python
x_other_diseases = []
y_other_diseases = []
for i in range(len(disease_freq)):
    if disease_freq[i] > 100 and disease_freq.index[i] not in ["ALL", "AML", "healthy"]: 
        x_other_diseases.append(disease_freq.index[i])
        y_other_diseases.append(disease_freq[i])
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x=x_other_diseases, y=y_other_diseases)
plt.savefig(parent + '/Figures/other_diseases.eps',
           dpi=300)
```

```python
x_aml_cll_all= []
y_aml_cll_all = []
for i in range(len(disease_freq)):
    if disease_freq[i] > 500 and disease_freq.index[i] not in ["healthy"]: 
        x_aml_cll_all.append(disease_freq.index[i])
        y_aml_cll_all.append(disease_freq[i])
```

```python
sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x=x_aml_cll_all, y=y_aml_cll_all)
plt.savefig(parent + '/Figures/aml_all_cll.eps',
           dpi=300)
```

```python
x_aml_healthy= []
y_aml_healthy = []
for i in range(len(disease_freq)):
    if disease_freq[i] > 800 and disease_freq.index[i] not in ["ALL"]: 
        x_aml_healthy.append(disease_freq.index[i])
        y_aml_healthy.append(disease_freq[i])

sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x=x_aml_healthy, y=y_aml_healthy)
plt.savefig(parent + '/Figures/aml_healthy.eps',
           dpi=300)
```

```python
# define Seaborn color palette to use
palette_color = sns.color_palette('deep')
  
# plotting data on chart
plt.pie(y_aml_healthy, labels=x_aml_healthy, colors=palette_color, autopct='%.0f%%', explode = [0,.2])
  
# displaying chart
plt.show()
```

```python
x_all_healthy= []
y_all_healthy = []
for i in range(len(disease_freq)):
    if disease_freq[i] > 800 and disease_freq.index[i] not in ["AML"]: 
        x_all_healthy.append(disease_freq.index[i])
        y_all_healthy.append(disease_freq[i])

sns.set(rc = {'figure.figsize':(15,8)})
sns.barplot(x=x_all_healthy, y=y_all_healthy)
plt.savefig(parent + '/Figures/all_healthy.eps',
           dpi=300)
```
