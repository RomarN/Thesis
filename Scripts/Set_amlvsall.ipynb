```python
import pandas as pd
random_state = 7
```

```python
df = pd.read_csv(r"C:\Users\RomarNathoeRepubliq\Documents\Uni\Thesis\Genes\Data\datasets123_allinfo.csv")
```

```python
df.head()
```

```python
print(df["Disease"].value_counts())
print()
```

```python
df = df.loc[df['Disease'].isin(['ALL','AML', 'AMKL'])]
```

```python
print(df["Disease"].value_counts())
print()
```

```python
df['target'] = [1 if (x == 'AML' or x == 'AMKL') else 0 for x in df['Disease']]
```

```python
df.head()
```

```python
print(df["target"].value_counts())
print()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.1, random_state=random_state, stratify=df[['target']])
```

```python
train.head()
```

```python
test.head()
```

```python
print(train["target"].value_counts())
print()
print(test["target"].value_counts())
print()
```

```python
train = train.drop(["Dataset", "GSE", "Condition", "Disease", "Tissue", "FAB", "Filename", "id"], axis = 1)
```

```python
test = test.drop(["Dataset", "GSE", "Condition", "Disease", "Tissue", "FAB", "Filename", "id"], axis = 1)
```

```python
train.to_csv("trainset_AMLvsALL_90_target_in.csv" ,index=False)
```
